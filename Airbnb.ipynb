**Filtering the Required Columns to Analyse**

```python
df_filtered = df[['_id',"accommodates",'neighborhood_overview','address.government_area','address.market','address.country','address.country_code','host.host_id','host.host_name','host.host_is_superhost','host.host_location','host.host_neighbourhood','host.host_response_time',"number_of_reviews",'guests_included',"reviews_per_month",'room_type','bed_type','minimum_nights','maximum_nights','address.location.coordinates[0]' ,'address.location.coordinates[1]','price','bedrooms','beds','property_type','number_of_reviews','first_review','availability.availability_30','availability.availability_60','availability.availability_90','availability.availability_365']]
```

```python
df_filtered.shape
```

```python
#Remnaming column names
df_filtered = df_filtered.rename({"neighborhood_overview":'neighbourhood',"address.location.type":"Location_type","host.host_location":"host_location","host.host_verifications[1]":"host_verification_2","host.host_verifications[0]":"host_verification_1","host.host_is_superhost":"host_is_superhost","host.host_name":"host_name","address.market":"city","address.government_area":'government_area',"address.suburb":"suburb","address.street":"street","images.thumbnail_url":"image","host.host_response_time":"host_response_time","_id ":"ID","host.host_id":"host_id","host.host_name ":"host_name","host.host_neighbourhood":"host_neighbourhood","address.country":"country","address.country_code":"country_code","availability.availability_30":"availability_30","availability.availability_60":"availability_60","availability.availability_90":"availability_90",'availability.availability_365':"availability_365","address.location.coordinates[0]":"Latitudes","address.location.coordinates[1]":"Longitudes"},axis="columns")
```

```python
df_filtered.info()
```

```python
# Eliminating Null Values
```

```python
df_filtered.isnull().sum()
```

```python
df_filtered["neighbourhood"].fillna(0, inplace=True)
df_filtered["host_neighbourhood"].fillna(0, inplace=True)
df_filtered["host_response_time"].fillna(0, inplace=True)
df_filtered["beds"].fillna(0, inplace=True)
df_filtered["bedrooms"].fillna(0, inplace=True)
df_filtered["city"].fillna(0, inplace=True)
df_filtered["first_review"].fillna(0, inplace=True)
df_filtered["host_location"].fillna(0, inplace=True)
df_filtered["host_location"].fillna(0, inplace=True)
df_filtered["reviews_per_month"].fillna(0, inplace=True)
```

```python
df_filtered.isnull().sum()
```

**Altering Data Types**

```python
df_filtered.dtypes
```

```python
# changing Data Types
df_filtered["price"]= df_filtered["price"].astype(str).astype(float)
df_filtered["bedrooms"]= df_filtered["bedrooms"].astype(str).astype(float).astype(int)
df_filtered["beds"]= df_filtered["beds"].astype(str).astype(float).astype(int)
df_filtered["host_response_time"].fillna("Not Specified",inplace= True)
df_filtered["reviews_per_month"]= df_filtered["reviews_per_month"].astype(str).astype(float).astype(int)
```

```python
df_filtered.shape
```

```python
df_filtered["country"].unique()
```

```python
#Changing Date Time format
df_filtered["first_review"]=pd.to_datetime(df_filtered["first_review"])
```

### **Filtering Amenities**

```python
df_amenities = df.iloc[0:5555,25:31]
```

```python
df_amenities.dtypes
```

```python
# Concatenate values from multiple columns into a single column separated by commas
df_amenities['Amenities'] = df_amenities[['amenities[0]', 'amenities[1]', 'amenities[2]', 'amenities[3]', 'amenities[4]', 'amenities[5]']].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)
```

```python
df_amenities
```

```python
df_amenities=df_amenities.drop(["amenities[0]","amenities[1]","amenities[2]","amenities[3]","amenities[4]","amenities[5]"],axis="columns")
```

```python
df_amenities = df_amenities.sort_values(by='Amenities')
```
